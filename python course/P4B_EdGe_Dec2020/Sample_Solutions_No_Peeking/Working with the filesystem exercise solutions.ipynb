## Binning DNA sequences

First, get a list of the files we need to process. We only want the ones ending in `.dna`.

```python
import os
for file_name in os.listdir("dna_files/"):
    if file_name.endswith(".dna"):
        print("reading sequences from " + file_name)
        
        
```

Now open each file and process each line. Remember to construct the path the file using the folder name plus the file name.

First thing is to calculate the length:

```python
# look at each file
for file_name in os.listdir("dna_files/"): 
    if file_name.endswith(".dna"): 
        print("reading sequences from " + file_name) 
        dna_file = open("dna_files/" + file_name)

        # look at each line 
        for line in dna_file: 
            dna = line.rstrip("\n") 
            length = len(dna) 
            print("found a dna sequence with length " + str(length)) 
```

Now try each bin in turn to see if the sequence fits. Use a range to check the bounds on each bin:

```python
# go through each file in the folder
for file_name in os.listdir("dna_files"): 

  # check if it ends with .dna
  if file_name.endswith(".dna"): 
    print("reading sequences from " + file_name) 

    # open the file and process each line
    dna_file = open("dna_files/" + file_name) 
    for line in dna_file:

      # calculate the sequence length 
      dna = line.rstrip("\n") 
      length = len(dna) 
      print("sequence length is " + str(length)) 

      # go through each bin and check if the sequence belongs in it
      for bin_lower in range(100,1000,100): 
        bin_upper = bin_lower + 99 
        if length >= bin_lower and length <= bin_upper: 
          print("bin is " + str(bin_lower) + " to " + str(bin_upper)) 
```

We can now write out each DNA sequence to the correct output file. We construct the file name based on the bin boundaries, and open the output file in append mode so that we don't overwrite earlier sequences. 

```python
import os
import shutil
 
# process all files that end in .dna
for file_name in os.listdir("dna_files/"): 
  if file_name.endswith(".dna"): 
    print("reading sequences from " + file_name) 
    dna_file = open("dna_files/" + file_name) 

    # for each line, calculate the sequence length
    for line in dna_file: 
      dna = line.rstrip("\n") 
      length = len(dna) 
      print("sequence length is " + str(length)) 

      # figure out which bin the sequence belongs in        
      for bin_lower in range(100,1000,100): 
        bin_upper = bin_lower + 99 
        if length >= bin_lower and length < bin_upper: 

          # once we know the correct bin, write out the sequence
          print("bin is " + str(bin_lower) + " to " + str(bin_upper)) 
          output_file_name = str(bin_lower) + "_" + str(bin_upper) + '.dna' 
          output = open(output_file_name, "a") 
          output.write(dna + "\n") 
          output.close() 
```

## Kmer counting

To figure out which kmers occur more than the threshold number of times, we have to add up __all__ the kmers for __all__ the sequences in __all__ the files before we know. So we need to build up the counts first, then process them. 

We'll start with a function to split a DNA sequence into kmers of a given length:

```python
def split_dna(dna, kmer_size): 
    kmers = [] 
    for start in range(0,len(dna)-(kmer_size-1),1): 
        kmer = dna[start:start+kmer_size] 
        kmers.append(kmer) 
    return kmers 
 
 
print(split_dna("AATGCTGCAT", 4)) 
```

Now we can plug this into the code from the previous exercise for processing all sequences in all files. We will use a dict to keep track of the counts:

```python
import os 
kmer_size = 6 
 
def split_dna(dna, kmer_size): 
    kmers = [] 
    for start in range(0,len(dna)-(kmer_size-1),1): 
        kmer = dna[start:start+kmer_size] 
        kmers.append(kmer) 
    return kmers 
 
kmer_counts = {} 
for file_name in os.listdir("dna_files/"): 
    if file_name.endswith(".dna"): 
        print("reading sequences from " + file_name) 
        dna_file = open("dna_files/" + file_name) 
        for line in dna_file: 
            dna = line.rstrip("\n") 
            for kmer in split_dna(dna, kmer_size): 
                current_count = kmer_counts.get(kmer, 0) 
                new_count = current_count + 1 
                kmer_counts[kmer] = new_count 
 
print(kmer_counts) 
```

We end up with a very big dict. Now we can go through the dict one item at a time and check if the kmer count is bigger than the cutoff:

```python
count_cutoff = 23 
for kmer, count in kmer_counts.items(): 
    if count > count_cutoff: 
        print(kmer + " : " + str(count)) 
```
